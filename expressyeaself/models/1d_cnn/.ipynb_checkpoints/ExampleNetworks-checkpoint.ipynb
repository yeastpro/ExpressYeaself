## Example Networks

Some general imports and setting up any directories that might be needed for the model

```python
%matplotlib inline
import pandas as pd
import numpy as np
import os, sys, subprocess
import itertools
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm

# ROOT_DIR - root directory
ROOT_DIR = os.getcwd()+'/'

# CHECKPOINTS_DIR  - directory where model checkpoints are automatically saved
CHECKPOINTS_DIR = ROOT_DIR + 'checkpoints/'
```

Read the input data file.

In this case I just shortened the name and dropped it into the root directory of my little example project

```python
DF_full = pd.read_csv(ROOT_DIR+'Abf1TATA.txt', sep="\t", header=None)
DF_full.columns = ['sequence','expression']
print ('Number of sequences:', len(DF_full))
```

Create a smaller data frame to use for the examples

```python
DF_small = DF_full[0:10000]
DF_small.shape
```

```python
max_sequence_len = DF_full['sequence'].apply(len).max()
print('Max Sequence Length:', max_sequence_len)
```

Setup a dictonary object to use for encoding letter sequence to numerical sequence

```python
all_nucleotides = list("ATGCNP")
df_na = pd.DataFrame.from_dict({'nucleotide':all_nucleotides})
df_na
```

Add numercal encoding and one_hot encoding to the example data frame.


I'm doing this using a little parallelization to speed things up. With the full dataset I might rework this using PyData (GPU processing)
otherwize it would take days to encode the 23M rows

```python
from joblib import Parallel, delayed
from tensorflow.python.keras.utils import to_categorical
import multiprocessing

def sequenceBuilder(i,row):
    row_seq = np.array(list(row['sequence']))
    row_na = np.array([df_na[df_na.nucleotide==na].index[0] for na in row_seq], dtype='int16') + 1
    row_one_hot = to_categorical(row_na, num_classes=len(df_na)+1, dtype='int16')
    return row_seq,row_na,row_one_hot

num_cores = multiprocessing.cpu_count()
result = Parallel(n_jobs=num_cores)(delayed(sequenceBuilder)(i,row) for i,row in tqdm(DF_small.iterrows()))

DF_small['na_index'] = pd.Series([el[1] for el in result], index= DF_small.index)
DF_small['na_one_hot'] = pd.Series([el[2] for el in result], index= DF_small.index)
```

Some additional deep learning imports and a handly little plot function

```python
import tensorflow as tf
from tensorflow.python.keras import backend as K
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.python.keras.preprocessing import sequence
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Input, Dense, Dropout, Embedding, LSTM, Conv1D, LocallyConnected1D, MaxPooling1D, GlobalAveragePooling1D
from tensorflow.python.keras import regularizers
from tensorflow.python.keras.utils import to_categorical
from tensorflow.python.keras.callbacks import ModelCheckpoint, TensorBoard
from sklearn.model_selection import StratifiedKFold

tb = TensorBoard(log_dir='./logs', histogram_freq=3, batch_size=batch_size, write_graph=True, write_grads=True, 
                 write_images=True)

def plot_results(hist):
    # summarize history for accuracy
    plt.subplot(1, 2, 1)
    plt.plot(hist['acc'])
    plt.plot(hist['val_acc'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')

    # summarize history for loss
    plt.subplot(1, 2, 2)
    plt.plot(hist['loss'])
    plt.plot(hist['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.tight_layout()
    
    return plt
```

Scale the expression data 

```python
scaler = MinMaxScaler()
y = DF_small['expression'].values

y_vals = y.reshape((len(y), 1))
scaler.fit(y_vals)
y_scaled = scaler.transform(y_vals)
```

### LSTM Model using one_hot encoding

```python
batch_size = 128

checkpointer = ModelCheckpoint(monitor='val_acc', filepath=CHECKPOINTS_DIR+'SimpleLSTM.hdf5', verbose=1, save_best_only=True)

X_padded = sequence.pad_sequences(DF_small['na_one_hot'].values,maxlen=max_sequence_len, padding='post')
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_scaled, test_size=0.3)

model = Sequential()
model.add(LSTM(100,input_shape=(max_sequence_len, 7)))
model.add(Dropout(0.25))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse',
              optimizer='rmsprop',
              metrics=['accuracy'])
print(model.summary())
# Fit the model
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=10,verbose=1,
                    validation_data=(X_test, y_test)) #callbacks=[checkpointer]

# evaluate the model
score = max(history.history['val_acc'])
print("%s: %.2f%%" % (model.metrics_names[1], score*100))
plt = plot_results(history.history)
plt.show()
```

### LSTM Model using embedding

```python
batch_size = 128

checkpointer = ModelCheckpoint(monitor='val_acc', filepath=CHECKPOINTS_DIR+'SimpleEmbdLSTM.hdf5', verbose=1, save_best_only=True)

X_padded = sequence.pad_sequences(DF_small['na_index'].values,maxlen=max_sequence_len, padding='post')
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_scaled, test_size=0.3)

model = Sequential()
model.add(Embedding(7, output_dim=8, input_length=max_sequence_len, mask_zero=True))
model.add(LSTM(100))
model.add(Dropout(0.25))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse',
              optimizer='rmsprop',
              metrics=['accuracy'])
print(model.summary())
# Fit the model
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=10,verbose=1,
                    validation_data=(X_test, y_test)) #callbacks=[checkpointer]

# evaluate the model
score = max(history.history['val_acc'])
print("%s: %.2f%%" % (model.metrics_names[1], score*100))
plt = plot_results(history.history)
plt.show()
```

### CNN Model using on-hot encoding

```python
batch_size = 128

checkpointer = ModelCheckpoint(monitor='val_acc', filepath=CHECKPOINTS_DIR+'SimpleEmbdLSTM.hdf5', verbose=1, save_best_only=True)

X_padded = sequence.pad_sequences(DF_small['na_one_hot'].values,maxlen=max_sequence_len, padding='post')
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_scaled, test_size=0.3)

model = Sequential()
model.add(Conv1D(32, 3, activation='relu', input_shape=(max_sequence_len, 7)))
model.add(Conv1D(32, 3, activation='relu'))
model.add(MaxPooling1D(3))
model.add(Conv1D(128, 5, activation='relu'))
model.add(Conv1D(128, 5, activation='relu'))
model.add(GlobalAveragePooling1D())
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mse',
              optimizer='rmsprop',
              metrics=['accuracy'])
print(model.summary())
# Fit the model
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=10,verbose=1,
                    validation_data=(X_test, y_test)) #callbacks=[checkpointer]

# evaluate the model
score = max(history.history['val_acc'])
print("%s: %.2f%%" % (model.metrics_names[1], score*100))
plt = plot_results(history.history)
plt.show()
```
